```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(
    "C:/D Drive Contents/Driver assist - Internship/Driver_monitoring/Recordings/Pradarshana_disp/data.csv"
)

# Loop through each row in the DataFrame
for index, row in df.iterrows():

    # Extract the values from the 3 columns of interest for the current row
    focused_value = row["Focused"]
    other_distract_value = row["Other_Distractions"]
    sleepy_value = row["Sleepy"]

    # Create a bar chart using the extracted values
    labels = ["Focused", "Other Distractions", "Sleepy"]
    values = [focused_value, other_distract_value, sleepy_value]

    # Set the figure size
    fig = plt.figure(figsize=(13.333, 15))

    # Create the bar chart
    plt.bar(labels, values, color=["g", "m", "r"])

    # Set the chart title
    bars = plt.title(f"Number of Frames", fontsize=28)

    # Set the y-axis limit
    plt.ylim(1, 16000)
    plt.yscale("log")

    col = ["g", "m", "r"]
    # plt.text( str(focused_value), color='red', fontweight='bold', fontsize=12, ha='center', va='bottom')
    for i, v in enumerate(values):
        plt.text(
            i,
            v,
            str(v),
            color=col[i],
            fontweight="bold",
            fontsize=30,
            ha="center",
            va="bottom",
        )

    # Set the y-axis ticks
    # plt.yticks(range(0, 17000, 1000))
    plt.xticks(fontsize=28)
    plt.yticks(fontsize=28)

    # Save the chart as an image file
    plt.savefig(f"C:/Users/manis/OneDrive/Desktop/bar_graph/{index + 1}.png")

    plt.close(fig)

    # Clear the current figure
    plt.clf()
```

```python
import cv2
import os
from moviepy.video.io.VideoFileClip import VideoFileClip

# Load the video file
clip = VideoFileClip("C:/Users/manis/Videos/Original_30_Updt.mp4")

# Get the duration of the video in seconds
duration = clip.duration

# Get number of frames
files = os.listdir("C:/Users/manis/OneDrive/Desktop/bar_plots")
frames = len(files)


# Set the input and output file paths
input_path = "C:/Users/manis/OneDrive/Desktop/bar_plots"
output_path = "C:/Users/manis/OneDrive/Desktop/video/bar_61.75.mp4"

# Get the list of image file names in the input directory
image_files = os.listdir(input_path)


"""print(image_files)"""
# Set the frame rate (in frames per second)
frame_rate = 61.75
print(duration, frames, frame_rate)

# Set the output video codec and create the video writer object
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
video_writer = cv2.VideoWriter(output_path, fourcc, frame_rate, (1333, 1500))

# Iterate over the sorted image files and add them to the video
for image_file in range(int(len(image_files))):
    # Load the image from file
    image_path = os.path.join(input_path, str(image_file) + ".png")
    print(image_path)
    image = cv2.imread(image_path)

    # Resize the image to match the output video resolution
    # image = cv2.resize(image, (1280, 720))

    # Write the image to the video
    video_writer.write(image)

# Release the video writer and destroy any open windows
video_writer.release()
cv2.destroyAllWindows()
```
