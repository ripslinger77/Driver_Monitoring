Generate video 

```python
import cv2
import os
from moviepy.video.io.VideoFileClip import VideoFileClip

# Load the video file
clip = VideoFileClip("C:/Users/manis/Videos/Iisc New 5Min.mp4")

# Get the duration of the video in seconds
duration = clip.duration

# Get number of frames
files = os.listdir("C:/Users/manis/OneDrive/Desktop/mapping_values")
frames = len(files) / 2


# Set the input and output file paths
input_path = "C:/Users/manis/OneDrive/Desktop/mapping_values"
output_path = "C:/Users/manis/OneDrive/Desktop/video/map.mp4"

# Get the list of image file names in the input directory
image_files = os.listdir(input_path)


"""print(image_files)"""
# Set the frame rate (in frames per second)
frame_rate = frames / duration

# Set the output video codec and create the video writer object
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
video_writer = cv2.VideoWriter(output_path, fourcc, frame_rate, (1280, 720))

# Iterate over the sorted image files and add them to the video
for image_file in range(int(len(image_files) / 2)):
    # Load the image from file
    image_path = os.path.join(input_path, str(image_file) + ".png")
    print(image_path)
    image = cv2.imread(image_path)

    # Resize the image to match the output video resolution
    image = cv2.resize(image, (1280, 720))

    # Write the image to the video
    video_writer.write(image)

# Release the video writer and destroy any open windows
video_writer.release()
cv2.destroyAllWindows()
```

```python
import folium
import pandas as pd

# read the CSV file into a pandas dataframe
df = pd.read_csv(
    "C:/D Drive Contents/Driver assist - Internship/Driver_monitoring/programs/Pradrshana_disp/gps_data.csv"
)

# extract the latitude and longitude columns
latitudes = df["latitude"].tolist()
longitudes = df["longitude"].tolist()

# create map object
m = folium.Map(location=[latitudes[0], longitudes[0]], zoom_start=13)


# add markers to the map
for lat, lon in zip(latitudes, longitudes):
    folium.CircleMarker(
        location=[lat, lon], radius=2, color="red", fill=True, fill_color="red"
    ).add_to(m)

# create map image and save it as a PNG file
m.save("C:/Users/manis/OneDrive/Desktop/video/map.html")
map_image = folium.Figure().add_child(m)
map_image.save("C:/Users/manis/OneDrive/Desktop/video/map.png")
```
